```python
from sqlalchemy import create_engine

# Replace YOUR_PASSWORD with your actual password
engine = create_engine('postgresql://postgres:Password@001@localhost:5432/sales_analysis_db')
```

```python
import pandas as pd

xls = pd.ExcelFile(r'C:\Users\charlie00\Downloads\Regional Sales.xlsx')

# Read all sheets
sales_orders = pd.read_excel(xls, 'Sales Orders')
customers = pd.read_excel(xls, 'Customers')
regions = pd.read_excel(xls, 'Regions')
state_regions = pd.read_excel(xls, 'State Regions')
products = pd.read_excel(xls, 'Products')
budgets = pd.read_excel(xls, '2017 Budgets')

# Upload each sheet as table in PostgreSQL
sales_orders.to_sql('sales_orders', engine, index=False, if_exists='replace')
customers.to_sql('customers', engine, index=False, if_exists='replace')
regions.to_sql('regions', engine, index=False, if_exists='replace')
state_regions.to_sql('state_regions', engine, index=False, if_exists='replace')
products.to_sql('products', engine, index=False, if_exists='replace')
budgets.to_sql('budgets_2017', engine, index=False, if_exists='replace')

print("All tables uploaded successfully to PostgreSQL!")
```
